```
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import json
```

```
url = 'https://www.who.int/news-room/q-a-detail/q-a-coronaviruses'
response = requests.get(url)
print(response)
if(response.ok):
  data = response.text
```

```
soup = BeautifulSoup(data, 'lxml')
```

```
questions = soup.find_all('div', class_='sf-accordion__trigger-panel')
```

```
qs = []
for element in questions:
  text_q = element.get_text()
  qs.append(text_q)
```

```
qs
```

```
answers = soup.find_all('div', class_='sf-accordion__content')
```

```
ans = []
for element in answers:
  text_a = element.get_text()
  ans.append(text_a)
```

```
ans
```

```
print(len(qs))
```

```
print(len(ans))
```

```
with open('qs.json', 'w') as f:
  json.dump(qs, f)

with open('ans.json', 'w') as f:
  json.dump(ans, f)
```
