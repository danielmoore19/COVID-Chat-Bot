Scrape the Euro CDC Q&A

```
from bs4 import BeautifulSoup
import requests
import json
import re
```

```
url = 'https://www.ecdc.europa.eu/en/covid-19/questions-answers'
response = requests.get(url)
if(response.ok):
  data = response.text
```

```
soup = BeautifulSoup(data, 'lxml')
tags = ['h3', 'strong']
```

```
questions = soup.select('h3 strong')
```

```
eu_questions = []
for element in questions:
  text_q = element.get_text()
  eu_questions.append(text_q)
```

Not bad. Just gotta ditch the numbering system.

```
answers = soup.select('p')
```

```
eu_answers =[]
```

```
eu_answers2 = []
for element in answers:
  text_a = element.get_text()
  eu_answers2.append(text_a)
```

```
for element in soup.find_all("p"):
  if "ct__type" not in str(element):
    eu_answers.append(element.text)
```
