```
import requests
import re
import json
from bs4 import BeautifulSoup
```

```
url = ('https://www.fda.gov/emergency-preparedness-and-response/coronavirus-disease-2019-covid-19/coronavirus-disease-2019-covid-19-frequently-asked-questions')
response = requests.get(url)
if(response.ok):
  data = response.text
```

```
soup = BeautifulSoup(data, 'lxml')
```

```
text = soup.find_all('div')
```

```
regexQ = re.compile('Q: |Q. ')
regexA = re.compile('A. ')
```

```
q_and_a = [x.text.strip() for x in soup.find_all('div', {'title': regexQ})]
```

```
replaced = []
for line in q_and_a:
  corrected = line.replace('A.', 'A:')
  replaced.append(corrected)
```

```
q_a_split = []
for line in replaced:
  splitter = re.split(('A\: ?'), line)
  q_a_split.append(splitter)
```

```
q_a_split.pop(0)
```

```
fda_questions, fda_answers = zip(*q_a_split)
```

```
def text_cleaner(text):
    text = re.sub('(Q: *|Q. *)', '', text)
    text = re.sub(r'--',' ',text)
    text = re.sub("[\[].*?[\]]", "", text)
    #text = re.sub(r"(\b|\s+\-?|^\-?)(\d+|\d*\.\d+)\b", " ", text)
    text = ' '.join(text.split())
    return text
```

```
clean_qs =[]
for question in fda_questions:
  text = text_cleaner(question)
  clean_qs.append(text)
```

```
clean_as = []
for answer in fda_answers:
  text = text_cleaner(answer)
  clean_as.append(text)
```

```
with open('clean_qs.json', 'w') as f:
  json.dump(clean_qs, f)
with open('clean_as.json', 'w') as f:
  json.dump(clean_as, f)
```
