Imports to put together all the different FAQs we have collected into a single dataframe.

```
import pandas as pd
import json
import re
```

Read in our JSONs. Realized later that I had not cleaned my WHO q and a banks.

```
with open('/content/drive/My Drive/Colab Datasets/COVID Chat Bot/cdc_questions.json', 'r') as f:
  cdc_questions = json.load(f)

with open('/content/drive/My Drive/Colab Datasets/COVID Chat Bot/cdc_answers.json', 'r') as f:
  cdc_answers = json.load(f)

with open('/content/drive/My Drive/Colab Datasets/COVID Chat Bot/fda_questions.json', 'r') as f:
  fda_questions = json.load(f)

with open('/content/drive/My Drive/Colab Datasets/COVID Chat Bot/fda_answers.json', 'r') as f:
  fda_answers = json.load(f)

with open('/content/drive/My Drive/Colab Datasets/COVID Chat Bot/who_questions.json', 'r') as f:
  messy_questions = json.load(f)

with open('/content/drive/My Drive/Colab Datasets/COVID Chat Bot/who_answers.json', 'r') as f:
  messy_answers = json.load(f)
```

Define our text cleaner with a function we've been using since the NLP module. Commented out the line that removes numbers because we need them.

```
def text_cleaner(text):
    text = re.sub(r'--',' ',text)
    text = re.sub("[\[].*?[\]]", "", text)
    #text = re.sub(r"(\b|\s+\-?|^\-?)(\d+|\d*\.\d+)\b", " ", text)
    text = ' '.join(text.split())
    return text
```

Clean up our questions and answers.

```
who_questions = []
for question in messy_questions:
  text = text_cleaner(question)
  who_questions.append(text)
```

```
who_answers = []
for answer in messy_answers:
  text = text_cleaner(answer)
  who_answers.append(text)
```

Zip our lists so we can make some frames.

```
cdc_list = list(zip(cdc_questions, cdc_answers))
fda_list = list(zip(fda_questions, fda_answers))
who_list = list(zip(who_questions, who_answers))
```

Make our frames and read in the one frame we already have.

```
cdc_df = pd.DataFrame(cdc_list, columns=['questions', 'answers'])
fda_df = pd.DataFrame(fda_list, columns=['questions', 'answers'])
who_df = pd.DataFrame(who_list, columns=['questions', 'answers'])
eu_df = pd.read_csv('/content/drive/My Drive/Colab Datasets/COVID Chat Bot/ecdc_qa.csv', names=['questions', 'answers'])
```

Append all the things!

```
df = cdc_df.append(fda_df, ignore_index=True)
```

```
df = df.append(who_df, ignore_index=True)
```

```
df = df.append(eu_df, ignore_index=True)
```

Save it to a file for our final bot!

```
df.to_csv('covid_df.csv')
```
